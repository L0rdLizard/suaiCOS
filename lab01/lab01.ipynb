```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal as sp
import math
```

```python
def cos_sum(x, k):
    tmp = 0
    for i in range(1, k):
        tmp += ((-8)/(np.pi**2*(2*i-1)**2))*np.cos(np.pi*(2*i-1)*x)
    return tmp
        

def sin_sum(x, k):
    tmp = 0
    for i in range(1, k):
        tmp += 2*((-((-1)**i)-1)/(np.pi*i))*np.sin(np.pi*i*x)
    return tmp

t = np.linspace(0, 1, 100)

fig, axs = plt.subplots(2, 4, sharey=False, figsize = (20, 10))
j = 0
for i in fig.axes:
    j+=1
    if j%4 == 0:
        i.plot(t, (2*t - 1))
    elif (j < 4):
        i.plot(t, np.array([cos_sum(t[iii], 10*(j + 1)) for iii in range(len(t))]))
    elif (j > 4):
        i.plot(t, np.array([sin_sum(t[iii], 20*(j + 1)) for iii in range(len(t))]))
```
